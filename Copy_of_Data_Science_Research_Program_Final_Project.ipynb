```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'flare_intensity': [100, 150, 80, 200, 170],
    'sunspot_area': [50, 65, 45, 80, 70],
    'xray_flux': [2.5, 3.0, 2.0, 3.5, 3.2]
}

df = pd.DataFrame(data)


print(df)


plt.plot(df['flare_intensity'], marker='o')
plt.title('Solar Flare Intensity Over Observations')
plt.xlabel('Observation Number')
plt.ylabel('Flare Intensity')
plt.show()

df.mean().plot(kind='bar')
plt.title('Average Solar Flare Features')
plt.xlabel('Feature')
plt.ylabel('Average Value')
plt.show()
```
