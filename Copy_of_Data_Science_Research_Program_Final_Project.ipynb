```python
import pandas as pd
import matplotlib.pyplot as plt

flare_data = {
    'flare_intensity': [100, 150, 80, 200, 170, 90, 120]
}
df = pd.DataFrame(flare_data)

plt.plot(df['flare_intensity'], marker='o')
plt.title('Solar Flare Intensity Over Observations')
plt.xlabel('Observation Number')
plt.ylabel('Flare Intensity')
plt.grid(True)
plt.show()
```
